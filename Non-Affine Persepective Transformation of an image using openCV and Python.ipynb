```python
src_pts=np.float32([[50,50],[450,40],[400,440],[40,440]])
```

```python
width=img.shape[0]
height=img.shape[1]
```

```python
destination=np.float32([[0,0],[width,0],[width,height],[0,height]])
```

```python
matrix= cv2.getPerspectiveTransform(src_pts, destination)
```

```python
img1= cv2.warpPerspective(img,matrix, (width, height))
```

```python
plt.imshow(img1)
```
